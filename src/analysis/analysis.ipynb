```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. S3から落としてきたCSVを読み込む
df = pd.read_csv('nba_metrics.csv')

# --- グラフ描画の魔法の設定 ---
plt.style.use('ggplot') # グラフをちょっとオシャレにする
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6)) # グラフを横に2つ並べる

# 📊 グラフ1: エンゲージメント率トップ10（棒グラフ）
# エンゲージメント率が高い順に並び替えて上から10個取得
top_10 = df.sort_values('engagement_rate', ascending=False).head(10)
# インデックス（動画番号）を文字列にして横軸にする
ax1.bar(top_10.index.astype(str), top_10['engagement_rate'], color='royalblue')
ax1.set_title('Top 10: Engagement Rate (%)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Video ID (Row Number)')
ax1.set_ylabel('Engagement Rate (%)')

# 📈 グラフ2: 再生数 vs エンゲージメント率（散布図）
ax2.scatter(df['playCount'], df['engagement_rate'], color='darkorange', alpha=0.7, s=100)
ax2.set_title('Play Count vs Engagement Rate', fontsize=14, fontweight='bold')
ax2.set_xlabel('Play Count (Views)')
ax2.set_ylabel('Engagement Rate (%)')

# グラフを表示！
plt.tight_layout()
plt.show()
```
